```python
from scipy import stats
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
import statsmodels.api as sm
input_dir = './'
```

```python
names = ['suicide', 'drug', 'poverty', 
#          'pollution', 
         'hunger', 
#          'literacy', 
         'rights', 
         'gdp']
file_names = ['suicide-death-rates.csv', 
              'share-with-alcohol-or-drug-use-disorders.csv',
#               'share-of-population-in-extreme-poverty.csv',
'share-deaths-air-pollution.csv',
'prevalence-of-undernourishment.csv',
# 'primary-completion-rate.csv',
'human-rights-protection.csv',
'gdp-per-capita-in-us-dollar-world-bank.csv',
             ]


col_of_interest = [
    "Deaths - Self-harm - Sex: Both - Age: Age-standardized (Rate)",
"Prevalence - Substance use disorders - Sex: Both - Age: Age-standardized (Percent)",
# '$1.90 per day - share of population below poverty line',
'Deaths - Cause: All causes - Risk: Air pollution - Sex: Both - Age: Age-standardized (Percent)',
'Prevalence of undernourishment (% of population)',
# 'Primary completion rate, total (% of relevant age group)',
'Human rights protection',
'GDP per capita (constant 2010 US$)',
]
```

```python
# create a base DF to add other ones to
df = pd.read_csv(input_dir+'suicide-death-rates.csv')
df = df[df.Year==2012]
df = df[~df['Code'].isnull()] #drop Nan
df = df[['Code', 'Year',col_of_interest[0]]] #keep only relevant cols, drop Entity since the country name may change from DF to DF
df
```

```python
for file, col in zip(file_names[1:], col_of_interest[1:]): #skip first DF about suicide
    df_i = pd.read_csv(input_dir+file)
    df_i = df_i[df_i.Year==2012]
    df_i = df_i[['Code', 'Year',col]] #keep only relevant cols, drop Entity since the country name may change from DF to DF
    print(file, df.shape)
    df = df.merge(df_i, on=['Code', 'Year'], how ='inner')
    

df
```

```python
df.to_csv(input_dir+'df_merged_150_countries.csv')
```

```python
df_150 = pd.read_csv('https://raw.githubusercontent.com/william-ma70/william-ma70.github.io/main/df_merged_150_countries.csv')
df_98 = pd.read_csv('https://raw.githubusercontent.com/william-ma70/william-ma70.github.io/main/df_merged_98_countries.csv')
df_160 = pd.read_csv('https://raw.githubusercontent.com/william-ma70/william-ma70.github.io/main/df_merged_160_countries.csv')
```

```python
s = df_98["Deaths - Self-harm - Sex: Both - Age: Age-standardized (Rate)"]
s_drug = df_98["Prevalence - Substance use disorders - Sex: Both - Age: Age-standardized (Percent)"]
s_poverty = df_98['$1.90 per day - share of population below poverty line']
s_pollution = df_98['Deaths - Cause: All causes - Risk: Air pollution - Sex: Both - Age: Age-standardized (Percent)']
s_hunger = df_98['Prevalence of undernourishment (% of population)']
s_literacy = df_98['Primary completion rate, total (% of relevant age group)']
s_rights = df_98['Human rights protection']
s_gdp = df_98['GDP per capita (constant 2010 US$)']

######### Evaluate the spearman correlation
print(stats.spearmanr(s, s_drug))
print(stats.spearmanr(s, s_poverty))
print(stats.spearmanr(s, s_pollution))
print(stats.spearmanr(s, s_hunger))
print(stats.spearmanr(s, s_literacy))
print(stats.spearmanr(s, s_rights))
print(stats.spearmanr(s, s_gdp))
```

```python
df_plot = pd.DataFrame({'Suicide Rates': s, 'Drug Use': s_drug})
corr = df_plot.corr(method = 'spearman')
sns.jointplot(x="Suicide Rates", y="Drug Use", data=df_plot, alpha=0.2)
plt.xlabel('xlabel', fontsize=24)
plt.ylabel('ylabel', fontsize=14)
# plt.savefig()
plt.show

df_plot = pd.DataFrame({'Suicide Rates': s, 'Poverty': s_poverty})
corr = df_plot.corr(method = 'spearman')
sns.jointplot(x="Suicide Rates", y="Poverty", data=df_plot, alpha=0.2)
plt.show()

df_plot = pd.DataFrame({'Suicide Rates': s, 'Pollution': s_pollution})
corr = df_plot.corr(method = 'spearman')
sns.jointplot(x="Suicide Rates", y="Pollution", data=df_plot, alpha=0.2)
plt.show()

df_plot = pd.DataFrame({'Suicide Rates': s, 'Hunger': s_hunger})
corr = df_plot.corr(method = 'spearman')
sns.jointplot(x="Suicide Rates", y="Hunger", data=df_plot, alpha=0.2)
plt.show()

df_plot = pd.DataFrame({'Suicide Rates': s, 'Literacy': s_literacy})
corr = df_plot.corr(method = 'spearman')
sns.jointplot(x="Suicide Rates", y="Literacy", data=df_plot, alpha=0.2)
plt.show()

df_plot = pd.DataFrame({'Suicide Rates': s, 'Rights': s_rights})
corr = df_plot.corr(method = 'spearman')
sns.jointplot(x="Suicide Rates", y="Rights", data=df_plot, alpha=0.2)
plt.show()

df_plot = pd.DataFrame({'Suicide Rates': s, 'GDP': s_gdp})
corr = df_plot.corr(method = 'spearman')
sns.jointplot(x="Suicide Rates", y="GDP", data=df_plot, alpha=0.2)
plt.show()
```

```python
from scipy.stats import spearmanr

def calculate_pvalues(df_160):
    df_160 = df_160.dropna()._get_numeric_data()
    dfcols = pd.DataFrame(columns=df_160.columns)
    pvalues = dfcols.transpose().join(dfcols, how='outer')
    for r in df_160.columns:
        for c in df_160.columns:
            pvalues[r][c] = spearmanr(df_160[r], df_160[c])[1]
    return pvalues


pvalues = calculate_pvalues(df_160.drop(['Year', 'Code'], axis=1))
pvalues
```

```python
sm.stats.fdrcorrection(pvalues, alpha=0.05, method='indep', is_sorted=False)
# mne.stats.fdr_correction(pvals, alpha=0.05, method='indep')
```
